```python
import pandas as pd

# Load the Titanic training data
train_data = pd.read_csv('train.csv')
print(train_data.head())
```

```python
train_data = train_data.drop(['PassengerId', 'Name','Ticket', 'Cabin'],axis =1)
print(train_data.head())
```

```python
train_data[['Age', 'Fare']] = train_data[['Age', 'Fare']].fillna(train_data[['Age', 'Fare']].mean())

# Fill missing categorical values (e.g., 'Embarked') with the mode
train_data['Embarked'] = train_data['Embarked'].fillna(train_data['Embarked'].mode()[0])

# Check if there are any missing values left
print(train_data.isnull().sum())
```

```python
train_data['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
train_data['Embarked'] = train_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
print(train_data.head())
```

```python
x= train_data.drop(columns= ['Survived'])
y = train_data['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter = 200)
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Make predictions on the test set
y_pred = model.predict(x_test) # Changed X_test to x_test

# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Display the confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Display the classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
